```python
import cv2

def crop(img):
  height, width, ch = img.shape
  #print (height, width)

  if(width > height):
    start = round((width - height)/2)
    crop_img = img[:height-1, start:start + height - 1]

  elif(height > width):
    start = round((height - width)/2)
    crop_img = img[start:start + width - 1, :width-1]

  else:
    crop_img = img

  height, width, ch = crop_img.shape
  #print (height, width)

  return crop_img
```

```python
def crappify(img, dest):
  res = cv2.blur(img,(15,15))

  mean = 0
  var = 10
  sigma = var ** 0.5
  gaussian = np.random.normal(mean, sigma, (400, 400, 3))
  img = cv2.resize(img, (100, 100), interpolation = cv2.INTER_NEAREST)
  img = cv2.resize(img, (400, 400), interpolation = cv2.INTER_NEAREST)
  img = img + gaussian
  cv2.imwrite(dest, img, [int(cv2.IMWRITE_JPEG_QUALITY), 50])
```

```python
from PIL import Image

def make_square(img):
# Take out the biggest and most central square from it that it can
    im = Image.open(img)
    x, y = im.size
    size = min(x, y)
    im = im.crop(((x - size) // 2, (y - size) // 2,
             (x + size) // 2, (y + size) // 2))
    #return im
    return numpy.array(im)
```
